```python
import pandas as pd
import hashlib
from google.colab import files
import io

# --- STEP 1: DEFINE FILE PATHS ---
# Assuming the files are already uploaded to the Colab environment
# and are in the /content/ directory. You might need to adjust these paths
# if your files are in a different location or have different names.
ib_filename = '/content/Install Base Original.xlsx'
map_filename = '/content/Install Base Mapping.xlsx'

print(f"Using Install Base file: {ib_filename}")
print(f"Using Mapping Table file: {map_filename}")

# --- STEP 2: PROCESS THE DATA ---
def process_colab_data(ib_path, map_path):
    # Load dataframes
    df_ib = pd.read_excel(ib_path)
    df_map = pd.read_excel(map_path)

    # Print column names for debugging
    print("\nInstall Base DataFrame Columns:", df_ib.columns.tolist())
    print("Mapping Table DataFrame Columns:", df_map.columns.tolist())

    # Standardize System Model
    # Correcting 'System Model' to 'SystemModel' based on user input and debug output
    df = pd.merge(df_ib, df_map, left_on='SystemModel', right_on='SystemModel', how='left')
    df['SystemModel'] = df['NewSystemModel'].fillna(df['SystemModel'])

    # Scramble System Serial (Anonymize)
    def quick_hash(val):
        if pd.isna(val): return "UNKNOWN"
        return hashlib.md5(str(val).encode()).hexdigest()[:10].upper()

    #df['System Serial'] = df['System Serial'].apply(quick_hash)

    # Final Cleanup
    # Modify cols_to_keep to retain the 'SystemModel' column
    cols_to_keep = [c for c in df.columns if c not in ['NewSystemModel']]
    df_final = df[cols_to_keep]

    # --- STEP 3: SAVE AND DOWNLOAD ---
    output_name = "processed_install_base.csv"
    df_final.to_csv(output_name, index=False)

    print(f"\n✅ Done! Processed {len(df_final)} rows.")
    print(f"Triggering download for: {output_name}")
    files.download(output_name)

# Run the processor
process_colab_data(ib_filename, map_filename)
```
